# Loading Data From File using listdir 

```python
# Open a file and Loading Data
path = "mini_newsgroups"
dirs = os.listdir( path )
dictionary = {}
# This would print all the main files in directories
for file in dirs:
    dictionary[file] = {}
    subdirs = os.listdir( path+'/'+file)    
    for subfile in subdirs:
        dictionary[file][subfile] = {}
        data = open(path+'/'+file +'/' +subfile)
        
        temp = []
        flag = False
        for i in data:
            if i.split(':')[0] == 'Lines' or flag:
                flag = True
                temp.append(i.strip('\n').strip(':').strip(':').strip('>').strip(' ').strip('?').split(' '))
       #Collecting data and fitting in dictionary
        
        for i in temp:
            for j in i:
                if j not in stopwords.words('english'):
                    if j in dictionary[file][subfile]:
                        dictionary[file][subfile][j] +=1
                    else :
                        dictionary[file][subfile][j] = 1
    print('read :' , file)            
```

# Training Data in Multinomial from Scratch

```python
# function for fitting Data || Dictionary for DATA
def fit(X_train, Y_train):
    result = {}
    class_values = set(Y_train)
    for current_class in class_values:
        result[current_class] = {}
        result["total_data"] = X_train.sum().sum()
        current_class_rows = (Y_train == current_class)
        X_train_current = X_train[current_class_rows]
        Y_train_current = Y_train[current_class_rows]
        result[current_class]["total_count"] = X_train_current.sum().sum()
        for j in X_train:
            result[current_class][j] = X_train_current[j].sum()
                            
    return result
```

```python
def predictSinglePoint(dictionary, x):
    classes = dictionary.keys()
    best_p = -1000
    best_class = -1
    first_run = True
    for current_class in classes:
        if (current_class == "total_data"):
            continue
        p_current_class = probability(dictionary, x, current_class)
        if (first_run or p_current_class > best_p):
            best_p = p_current_class
            best_class = current_class
        first_run = False
    return best_class
```
