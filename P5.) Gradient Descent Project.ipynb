```python
import numpy as np
def step_gradient_full(data , learning_rate , m ):
    N = len(data)
    m_slope = np.zeros(data.shape[1]-1)
    for i in range(N):
        x = data[i,:-1]
        y = data[i,-1]
        for j in range(data.shape[1]-1):
            m_slope[j] += (-2/N) * (y- (x*m).sum())*data[i,j]
    
    new_m = np.zeros(data.shape[1]-1)

    for i in range(len(m)):
        new_m[i] = m[i] - (learning_rate * m_slope[i])
    
    return new_m 

def gd_full(data , learning_rate , num_iteration):
    m = np.zeros(data.shape[1]-1)
    
    for i in range(num_iteration):
        m = step_gradient_full(data , learning_rate , m )
        print(i , "Cost " , cost_full(data,m))
    return m

def cost_full(data , m):
    total_cost = 0
    M = len(data)
    for i in range(M):
        x = data[i,:-1]
        y = data[i,-1]
        total_cost += (1/M) * ((y-(x*m).sum())**2)
    return total_cost

def run_full():
    data = np.loadtxt("train_cycle_C.csv" ,  delimiter=',')
    learning_rate = 0.0000005
    num_iteration = 15
    m = gd_full(data , learning_rate , num_iteration)
    print(m)
    return m

h = run_full()
```
